```python
import os
import pandas as pd
from dotenv import load_dotenv
from binance import Client, ThreadedWebsocketManager, ThreadedDepthCacheManager
import datetime as dt
```

```python
load_dotenv()
api_key = os.getenv("binance_api")
```

```python
client = Client(api_key=api_key)
```

```python
symbol = "BTCAUD"
interval="1m"

#Client.KLINE_INTERVAL_15MINUTE 
klines = client.get_historical_klines(symbol, interval, "6 Jun, 2022")
btc_df = pd.DataFrame(klines)

 # create colums name
btc_df.columns = ['open_time','open', 'high', 'low', 'close', 'volume','close_time', 'qav','num_trades','taker_base_vol','taker_quote_vol', 'ignore']
            
# change the timestamp
btc_df.index = [dt.datetime.fromtimestamp(x/1000.0) for x in btc_df.close_time]


#convert data to float and plot
btc_df = btc_df.astype(float)
btc_df = btc_df.reset_index()
btc_df = btc_df.rename(columns = {'index':'date'})
btc_df = btc_df[['date', 'close']]
btc_df.to_csv('../datasets/btc_df_Min.csv', index = None, header=True) 

```
