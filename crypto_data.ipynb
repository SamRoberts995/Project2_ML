```python
import os
import pandas as pd
from dotenv import load_dotenv
from binance import Client, ThreadedWebsocketManager, ThreadedDepthCacheManager
import datetime as dt
```

```python
load_dotenv()
api_key = os.getenv("binance_api")
```

```python
client = Client(api_key=api_key)
```

```python
# fetch daily kline
start = dt.datetime(2020,1,1)
end = dt.datetime.now()
timestamp = client._get_earliest_valid_timestamp('BTCAUD', '1d')
print(timestamp)
```

```python
datetime = dt.datetime.fromtimestamp(1596153600000/1000)
print(datetime)
```

```python
btc = client.get_historical_klines('BTCAUD', '1d', timestamp, limit=1000)
for line in btc:
    del line[5:]
btc_df = pd.DataFrame(btc, columns=['date', 'open', 'high', 'low', 'close'])
btc_df['date'] = btc_df['date'].apply(lambda ts: dt.datetime.fromtimestamp(ts/1000))
btc_df.set_index('date', inplace=True)
print(btc_df.head())
```
